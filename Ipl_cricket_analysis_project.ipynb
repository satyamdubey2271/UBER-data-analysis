```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
data = pd.read_csv("IPL 2025.csv")
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
figure = px.bar(data['Match_Winner'],color=data['Match_Winner'],title = "number of matches won by each teams in ipl 2008-2025")
figure.show()
```

```python
figure = px.bar(data['Player_of_Match'],color=data['Player_of_Match'],title = "best player of the match in IPL 2023")
figure.show()
```

```python
figure = px.scatter(data,x ='Middle_Overs_Scores',y ='Match_Winner',color=data['Middle_Overs_Scores'],title = "which team score the best middle over score in ipl 2023")
figure.show()
```

```python
figure = px.line(data,x='Powerplay_Scores',y='Match_Winner',color = data['Powerplay_Scores'],title
                 = 'which team has scores highest powerplay score in ipl 2023')
figure.show()
```

```python
figure = px.bar(data,x='Death_Overs_Scores',y='Match_Winner',color=data['Death_Overs_Scores'],title='which team make a highest death overs score')
figure.show()
```

```python
figure = px.scatter(data,x='Teams',y='Toss_Winner',color=data['Toss_Winner'],title=' toss winners  in ipl 2023 matches')
figure.show()
```

```python
figure = px.line(data,x='Player_of_Match',y='Teams',color=data['Player_of_Match'],title='player of the match in ipl 2008-2025 for per match')
figure.show()
```
